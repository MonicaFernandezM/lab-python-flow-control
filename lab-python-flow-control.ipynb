```python
#Define a list called products
products = ['t-shirt', 'mug', 'hat', 'book', 'keychain']
#Create a empty dictionary 
inventory = {}
```

```python
#ask the user the quantity of each product 
for product in products:
    quantity = int(input(f"Enter the quantity for {product}: "))
    inventory[product] = quantity
```

```python
#Create an empty dictionary customer_orders
customer_orders = set()

#Ask name of three products 
while len(customer_orders) < 3:
    product = input("Enter a product to order:" )
    if product in products:
        customer_orders.add(product)
    else:
        print("Product not available. Choose from: ", products)
```

```python
#Print the products in the customers_order
print(f"Customers order: {customer_orders}")
```

```python
#Calcualte statistics
total_products_ordered = len(customer_orders)
percentage_products_ordered = (total_products_ordered / len(products)) * 100
order_status = (total_products_ordered, percentage_products_ordered)
```

```python
#Print the order 
print("Order Statistics: ")
print(f"Total Products Ordered: {total_products_ordered}")
print(f"Percentage of Products Ordered: {percentage_products_ordered}")
```

```python
#Update the inventory subtracting -1 
for product in customer_orders:
    inventory[product] -= 1
```

```python
#Print the inventory, displaying the quantity of each product 
print("Updated Inventory: ")
for product, quantity in inventory.items():
    print(f"{product}: {quantity}")
```
